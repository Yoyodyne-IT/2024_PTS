# Lookyloo

## Capture

-> https://lookyloo.circl.lu/capture

Can send one or more URLs, or any file that can be rendered in a browser (HTML document, most of the time).

We can send many settings:
* User agent (will impact the browser used for the capture)
* HTTP headers
* HTTP basic auth login/passowrd
* Proxy (http / socks5) - the proxy has to be accessible by the machine where the capture is done
* Cookies
* Browser settings: timezone, locale, geolocation, prefered color scheme

They are passed to an actual browser, and a few actions are triggered on the rendered page to make it look like a legitimate user. This involves mouse move, keyboard action, scroll, and attempts to wait until there are no network traffic anymore.

------------ 

The settings can be used for reproducing captures over time with a clean and controlled environment, this is very hard to do in a normal browser used for other things.

-----------

In Lookyloo context, the output of a capture is as follows:
* HTTP Archive of all the requests/responses
* The rendered page in the browser
* The cookie jar (that can be passed to another capture)
* Screenshot of the full page (if possible0
* Potential favicons (if possible)
* Downloads triggered when rendering the page

## Tree

Built out of the HAR file, and other metadata from the content of the page. Helps understanding what is happening when a URL is loaded by a browser to the point the landing page is reached, but also what the landing page does. 

* Nodes -> aggrerations of URLs with the same hostname that are triggered by the same parent in the tree
* Icons -> type of content on the nodes
* Subsequent captures -> use the cookie jar & sets the referer to the URL of the rendered page for the new capture
* Submit to 3rd party services
* Lookup against 3rd party services

## Pivoting from trees to other trees

With one capture, we gather a lot of data, and many of them can be used to find other captures with similar indicators.

The indicators can be related to the hoster (IPs, AS), to the domain(s), to the resources loaded to render the page, or even the content of the rendered page itself.

# Pandora

Usecase is somewhat narrow, but important to keep in mind:

* a user receives a file (USB/Mail)
* user's day job is not to investigate the file but they need to see the content of the file, or at least have an idea what it contains
* they often don't need to open it on their machine once they see what's inside 

Bonus: if the file contains a URL, they want to see what this URL contains

Entry point for submitting files:

* webform to upload a file from the disk
* dedicated mailbox where emails can be forwarded to (replies with a link)
* API, so ither tools can submit files to pandora and get a permalink (ie. Lookyloo)

## Usage

A bunch of modules are run on the file and they generally search for potentially malicious content in the file, and/or if the file is an executable.

It will trigger plenty of false positive and flag legitimate files as suspicious **but** it is expected as it also shows a screenshot and the users are stringly invited to never open the file on their machine.
